### Implementation

```python
def calc_perimeter(mat, init_r, real_r, init_c, real_c):
  
  # Retrieve submatrix
  submat = mat[init_r:init_r+real_r,init_c:init_c+real_c]
  
  # Sum the perimeter
  left = submat[:,0]
  right = submat[:,-1]
  top = submat[0,:]
  bottom = submat[-1,:]
  sum = np.sum(left) + np.sum(right) + np.sum(top) + np.sum(bottom)
  
  return sum
```

```python
# Establish limit of rows and columns
rows = m_shape[0]
cols = m_shape[1]

# Initialize empty lists
init_row = []
init_col = []
m = []
n = []

for r in range(rows):
  for c in range(cols):

    # Python begins indexing at 0
    real_r = r + 1
    real_c = c + 1

    # Show submatrix size
    print(r,"x",c)

    # Traverse the matrix
    for init_r in range(rows):
      for init_c in range(cols):

        # Python begins indexing at 0
        real_init_r = init_r + 1
        real_init_c = init_c + 1

        # Determining available space in the matrix from starting point
        space_rows = rows - init_r
        space_cols = cols - init_c

        # If there's enough space
        if real_r <= space_rows and real_c <= space_cols:
          
          # Case when matrix is 1x1
          if real_r == 1 and real_c == 1:
            val = mat[init_r,init_c]

          # Case when matrix is larger than 1x1
          else:
            val = calc_perimeter(mat, init_r, real_r, init_c, real_c)

          if val == 0:
            init_row.append(real_init_r)
            init_col.append(real_init_c)
            m.append(real_r)
            n.append(real_c)
        
        # Ignoring over-sized matrices for starting point
        else:
          continue
```

```python
print(init_row)
print(init_col)
print(m)
print(n)
```

```python
# Empty list for sizes
sizes = []

# Calculate the size of the matrices
for i in range(len(m)):
  m_val = m[i]
  n_val = n[i]
  mat_size = m_val*n_val
  sizes.append(mat_size)
```

```python
# Find the largest matrix
largest = max(sizes)
largest_idx = sizes.index(largest)
print("Largest matrix:", largest)
print("Index:", largest_idx)
```

```python
IR = init_row[largest_idx]
IC = init_col[largest_idx]
M = m[largest_idx]
N = n[largest_idx]
print("The largest matrix whose perimeter equals 0 is a",M,"x",N,"matrix starting at [",IR,",",IC,"] (normal notation, not Python)")
```
